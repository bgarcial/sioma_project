```python
# ylabels for classification
ylabels = []
"""
0: Pequeño (0-15)
1: Mediano (15-23)
2: Grande (23-)
"""

for i in range(n):
    if data[ylabel].iloc[i] < 15:
        ylabels.append(0)
    elif data[ylabel].iloc[i]>=15 and data[ylabel].iloc[i]<23:
        ylabels.append(1)
    elif data[ylabel].iloc[i]>=23: 
        ylabels.append(2)
```

```python
ylabels


# Convert to pandas
ylabels = pd.DataFrame(ylabels)

# Split y
y_e = ylabels.iloc[ind_e]
y_v = ylabels.iloc[ind_v]
y_t = ylabels.iloc[ind_t]

y_e.head()
```

```python
# Import Tree Classifier
from sklearn.tree import DecisionTreeClassifier

# Define a new instance
mtree = DecisionTreeClassifier(random_state=10000)

# Fit training data
try:
    clabels.remove('Direccion')
except:
    None
    
try:
    qlabels.remove('peso')
except:
    None
    
# Variables for tree
vt = qlabels + clabels
data_e[vt].columns

mtree.fit(data_e[vt], y_e)
```

```python
mtree.min_samples_leaf
```

```python
# Predict data
mtree.predict(data_e[vt])

mtree.max_depth

imp = mtree.feature_importances_
j = imp.argsort()

# Table of features
tft = []

for i in range(len(vt)):
    vname = vt[j[i]]
    importance = imp[j[i]]
    tft.append([vname, importance])
    
tft = pd.DataFrame(tft)
tft.head()
```

```python
mtree.score(data_e[vt], y_e)
```

```python
# Perform validation
from sklearn.model_selection import cross_val_score
mtreev = cross_val_score(mtree, data_v[vt], y_v, cv=30)

mtreev = pd.DataFrame(mtreev)
mtreev.describe()
```
