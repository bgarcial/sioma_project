```python
"""Perform Cluster Analysis using MiniBatch KMeans"""
```

```python
# Import required modules
from sklearn.cluster import MiniBatchKMeans
import numpy as np
import random
from sklearn import cluster
import itertools
import numpy as np
from joblib import Parallel, delayed
import multiprocessing
# Set a seed
random.seed(10000)
```

```python
 #qlabels.remove(ylabel)
nv = nq - 1;
    
# Perform anlysis using only two variables
evals2 = list(itertools.combinations(range(nv),2))
    
print(len(evals2))
    
# Perform analysis using only three variables
evals3 = list(itertools.combinations(range(nv),3))
    
print(len(evals3))
```

```python
# Define models of to evaluate

def run_models2(k, data, v, qlabels):
    v1 = qlabels[v[0]]
    v2 = qlabels[v[1]]
    model = cluster.MiniBatchKMeans(n_clusters=k, random_state=10000)
    model.fit(data[[v1, v2]])
    return [model, v, model.inertia_]
    
def run_models3(k, data, v, qlabels):
    model = cluster.MiniBatchKMeans(n_clusters=k, random_state=10000)
    v1 = qlabels[v[0]]
    v2 = qlabels[v[1]]
    v3 = qlabels[v[2]]
    model.fit(data[[v1, v2, v3]])
    return [model, v, model.inertia_]
```

```python
# Find best configurations for two variables
num_cores = multiprocessing.cpu_count()
k = 3
results2 = Parallel(n_jobs=num_cores)(delayed(run_models2)(k, datan_e, i, qlabels) for i in evals2)
    
print(results2)
    
inertias2 = [r[2] for r in results2]
print(inertias2)
    
c = 3;
ord_inertias2 = np.argsort(inertias2)
ind_inertias2 = ord_inertias2[0:c]
    
print(ord_inertias2[0:c])
```

```python
# Find best configurations for three parameters
results3 = Parallel(n_jobs=num_cores)(delayed(run_models3)(k, datan_e, i, qlabels) for i in evals3)

print(results3)
    
inertias3 = [r[2] for r in results3]
print(inertias3)
    
c = 3;
ord_inertias3 = np.argsort(inertias3)
ind_inertias3 = ord_inertias3[0:c]
print(ord_inertias3[0:c])
```

```python
# Tune up hyperparameters
from sklearn.model_selection import GridSearchCV
hyperparameters = dict(n_init=range(1,11), random_state=[10000])
    
def eval_models(data, qlabels, results, ind, hyperparameters, nvars):
    model = results[ind][0]
    clf = GridSearchCV(model, hyperparameters, cv=5, verbose=0)
        
    v1 = qlabels[results[ind][1][0]]
    v2 = qlabels[results[ind][1][1]]
        
    if nvars == 2:
        best_model = clf.fit(data[[v1, v2]])
    elif nvars == 3:
        v3 = qlabels[results[ind][1][2]]
        best_model = clf.fit(data[[v1, v2, v3]])
            
    return [best_model.estimator, results[ind][1], best_model.estimator.inertia_]
        
models2 = Parallel(n_jobs=num_cores)(delayed(eval_models)(datan_v, qlabels, results2, i, hyperparameters, 2) for i in ind_inertias2)
inerts2 = [m[2] for m in models2]
ind_model2 = np.argsort(inerts2)[0]
model2 = models2[ind_model2][0]
    
models3 = Parallel(n_jobs=num_cores)(delayed(eval_models)(datan_v, qlabels, results3, i, hyperparameters, 3) for i in ind_inertias3)
inerts3 = [m[2] for m in models3]
ind_model3 = np.argsort(inerts3)[0]
model3 = models3[ind_model3][0]
```

```python
# Use Test data
inertias_m1 = []
inertias_m2 = []
inertias_m3 = []
inertias_m4 = []
inertias_m5 = []
    
    
for i in range(1, 11):
    m1 = cluster.MiniBatchKMeans(n_clusters = i, random_state=10000)
    m1.fit(datan_e[qlabels]);        
    inertias_m1.append(m1.inertia_)
        
    m2 = cluster.MiniBatchKMeans(n_clusters = i, random_state=10000)
    m2.fit(datan_e[[qlabels[models2[ind_model2][1][0]],qlabels[models2[ind_model2][1][1]]]]);        
    inertias_m2.append(m2.inertia_)
        
    m3 = cluster.MiniBatchKMeans(n_clusters = i, random_state=10000)
    v31=models3[ind_model3][1][0]
    v32=models3[ind_model3][1][1]
    v33=models3[ind_model3][1][2]
    m3.fit(datan_e[[qlabels[v31],qlabels[v32], qlabels[v33]]]);        
    inertias_m3.append(m3.inertia_)
    
    m4 = cluster.MiniBatchKMeans(n_clusters = i, random_state=10000)
    m4.fit(datapca2_e);        
    inertias_m4.append(m4.inertia_)
        
    m5 = cluster.MiniBatchKMeans(n_clusters = i, random_state=10000)
    m5.fit(datapca3_e);        
    inertias_m5.append(m5.inertia_)
        
    if i == 3:
        labels_m1 = m1.predict(datan_t[qlabels])
        labels_m2 = m2.predict(datan_t[[qlabels[models2[ind_model2][1][0]],qlabels[models2[ind_model2][1][1]]]]);        
        labels_m3 = m3.predict(datan_t[[qlabels[v31],qlabels[v32], qlabels[v33]]]);
        labels_m4 = m4.predict(datapca2_t);
        labels_m5 = m5.predict(datapca3_t); 
```
