# 2. Forming the whole Sioma testing dataset - January 12th to August 3rd - All input variables

We will make a single dataset where all the standardized training data are from:

- Wind Direction
- Luminosity
- Precipitations
- Temperature
- Phreatic Level
- Wind Speed
- Cluster weight of the week from July 27th to August 3rd.

```python
sioma_28_weeks_testing = pd.concat([luminosity_testing, precipitations_testing, temperature_testing, wind_speed_testing, wind_direction_testing, phreatic_level_testing, racimitos_weight_testing], axis=1)
sioma_28_weeks_testing
```

## 2.1 Filling **`NaN:`** values
---

We evaluate if `sioma_28_weeks_testing` dataframe has null type `NaN`

Have null values.

```python
imputer = Imputer(missing_values = 'NaN', strategy = 'mean', axis = 0)

# We adjust the values transforming the columns where you need data 
# with the fit() method. 
# We pass the dataframe sioma_28_weeks_testing and 
# with the variables where I can find the empty values, in this case 
# all variables [::] and fit 
imputer = imputer.fit(sioma_28_weeks_testing[::])
sioma_testing_full_values = imputer.transform(sioma_28_weeks_testing[::])
```

```python
sioma_testing_full_values
```

```python
print(sioma_testing_full_values.shape)
np.isnan(sioma_testing_full_values)
```

```python
cols = list(sioma_28_weeks_testing.columns)
print(sioma_28_weeks_testing.shape)
print(sioma_testing_full_values.shape)
print(cols)
```

```python
# We turn sioma_testing_full_values to dataframe
sioma_28_weeks_testing_df = pd.DataFrame(data=sioma_testing_full_values, columns=cols)
print(sioma_28_weeks_testing_df.isnull().any())
sioma_28_weeks_testing_df.head(10)
```

```python
sioma_28_weeks_testing_df.to_csv('../data/processed/sioma-inputs-output-unified_testing_28_weeks.csv', sep=',', header=True, index=False)
sioma_28_weeks_testing_df.shape
```
