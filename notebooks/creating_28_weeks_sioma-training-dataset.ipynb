## 2.1 Filling **`NaN:`** values
---

We evaluate if `sioma_28_weeks` dataframe has null type `NaN`

Have null values.

We see that many columns have remained with null values, product of the concatenation of these different dataframes, since not all had the same
number of records `X[i]`

We proceed to fill in that missing data
A more common solution alternative is to take the average of the column values where the data is missing, in this case find the average of all the columns that our data set has `sioma_28_weeks`

As we said before, we will use as a strategy, fill in the missing data with the average of the other values in the column where data is missing.

axis is the axis along which we are going to do the Imputer operation, by default it is equal to 0, which means that this operation will be along the columns

```python
imputer = Imputer(missing_values = 'NaN', strategy = 'mean', axis = 0)

# We adjust the values transforming the columns where you need data 
# with the fit() method. 
# We pass the dataframe sioma_28_weeks and 
# with the variables where I can find the empty values, in this case 
# all variables [::] and fit 
imputer = imputer.fit(sioma_28_weeks[::])
sioma_training_full_values = imputer.transform(sioma_28_weeks[::])
```

The previous step generate the ** `sioma_training_full_values` ** array which don't have null values

```python
sioma_training_full_values
```

```python
print(sioma_training_full_values.shape)
np.isnan(sioma_training_full_values)
```

```python
cols = list(sioma_28_weeks.columns)
print(sioma_28_weeks.shape)
print(sioma_training_full_values.shape)
print(cols)
```

```python
# Then, the `sioma_training_full_values` don't have missing values
np.isnan(sioma_training_full_values)
```

```python
# We turn sioma_training_full_values to dataframe
sioma_28_weeks_training_df = pd.DataFrame(data=sioma_training_full_values, columns=cols)
print(sioma_28_weeks_training_df.isnull().any())
```
