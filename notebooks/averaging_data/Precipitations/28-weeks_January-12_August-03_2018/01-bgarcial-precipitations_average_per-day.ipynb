```python
from __future__ import unicode_literals

import pandas as pd
import numpy as np
```

```python
precipitations_28w_df = pd.read_csv('../../../../data/raw/FincaPorvenir/Metereologico/28-weeks_January-12_August-03_2018/' \
                                 'Precipitaciones_January-12_August-03_2018.csv', )
```

```python
print(precipitations_28w_df.shape)
precipitations_28w_df.head()
```

- We going to generate in only one dataframe one record per day of precipitations values, performing an average of the values per day.

```python
# We turn to datetime the Fecha column values 
# https://pandas.pydata.org/pandas-docs/version/0.20/generated/pandas.to_datetime.html
precipitations_28w_df['Fecha:'] = pd.to_datetime(precipitations_28w_df['Fecha:'])

# http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.resample.html
# We use resample() function to use the data like time series per day and we pass the
# Fecha: column at the on parameter and we add the Precipitación (ml) column which we add the 
# average of each one

precipitations_28w_average_per_day = precipitations_28w_df.resample('D', on='Fecha:')['Precipitación (ml)'].mean().reset_index()
```

```python
print(precipitations_28w_average_per_day.shape)
precipitations_28w_average_per_day
```

```python
# We export the precipitations_28w_average_per_day dataframe to .csv file
precipitations_28w_average_per_day.to_csv('../../../../data/interim/Precipitations/28-weeks_January-12_August-03_2018/' +'\n' 
                                  'Precipitations_Average_per-day_January-12_August-03.csv', sep=',', header=True, index=False)
```
