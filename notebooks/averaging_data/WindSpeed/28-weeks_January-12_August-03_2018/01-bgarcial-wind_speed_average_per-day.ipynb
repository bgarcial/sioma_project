```python
from __future__ import unicode_literals

import pandas as pd
import numpy as np
```

```python
wind_speed_28w_df = pd.read_csv('../../../../data/raw/FincaPorvenir/Metereologico/28-weeks_January-12_August-03_2018/' \
                                 'Velocidad-del-viento_January-12_August-03_2018.csv', )
```

```python
print(wind_speed_28w_df.shape)
wind_speed_28w_df.head()
```

```python
# We turn to datetime the Fecha column values 
# https://pandas.pydata.org/pandas-docs/version/0.20/generated/pandas.to_datetime.html
wind_speed_28w_df['Fecha:'] = pd.to_datetime(wind_speed_28w_df['Fecha:'])

# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.groupby.html?highlight=groupby#pandas.DataFrame.groupby
# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Grouper.html
# We group the dataframe wind_speed_28w_df the column Fecha: (se la decimos con Grouper)
# with a frecuency of D calendar days generating to each day the wind speed average value
wind_speed_28w_average_per_day = (wind_speed_28w_df.groupby(pd.Grouper(key='Fecha:', freq='D'))['Velocidad del viento (Km/h)'].mean().reset_index())
```

```python
print(wind_speed_28w_average_per_day.shape)
wind_speed_28w_average_per_day
```

```python
# We export the wind_speed_28w_average_per_day dataframe to .csv file
wind_speed_28w_average_per_day.to_csv('../../../../data/interim/WindSpeed/28-weeks_January-12_August-03_2018/' +'\n' 
                                  'Wind-Speed_Average_per-day_January-12_August-03.csv', sep=',', header=True, index=False)
```
