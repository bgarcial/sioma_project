```python
from __future__ import unicode_literals

import pandas as pd
import numpy as np
```

```python
temperature_28w_df = pd.read_csv('../../../../data/raw/FincaPorvenir/Metereologico/28-weeks_January-12_August-03_2018/' \
                                 'Temperatura_January-12_August-03_2018.csv', )
```

```python
print(temperature_28w_df.shape)
temperature_28w_df.head()
```

```python
# We turn to datetime the Fecha column values 
# https://pandas.pydata.org/pandas-docs/version/0.20/generated/pandas.to_datetime.html
temperature_28w_df['Fecha:'] = pd.to_datetime(temperature_28w_df['Fecha:'])

# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.groupby.html?highlight=groupby#pandas.DataFrame.groupby
# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Grouper.html
# We group the dataframe temperature_28w_df the column Fecha: (se la decimos con Grouper)
# with a frecuency of D calendar days generating to each day the temperature average value
temperature_28w_average_per_day = (temperature_28w_df.groupby(pd.Grouper(key='Fecha:', freq='D'))['Temperatura (C)'].mean().reset_index())
```

```python
print(temperature_28w_average_per_day.shape)
temperature_28w_average_per_day
```

```python
# We export the temperature_28w_average_per_day dataframe to .csv file
temperature_28w_average_per_day.to_csv('../../../../data/interim/Temperature/28-weeks_January-12_August-03_2018/' +'\n' 
                                  'Temperature_Average_per-day_January-12_August-03.csv', sep=',', header=True, index=False)
```
