```python
from __future__ import unicode_literals

import pandas as pd
import numpy as np
```

```python
temperature_l12w_df = pd.read_csv('../../../../data/raw/FincaPorvenir/Metereologico/Latest_12-weeks_May-4_August-03_2018/' \
                                 'Temperatura_May-4_August-03_2018.csv', )
```

```python
print(temperature_l12w_df.shape)
temperature_l12w_df.head()
```

```python
temperature_l12w_df['Fecha:'] = pd.to_datetime(temperature_l12w_df['Fecha:'])

# We set like index the Fecha: column 
# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.set_index.html
temperature_l12w_df = temperature_l12w_df.set_index('Fecha:')

# http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.resample.html
# We use resample() function to use the data like time series per day and we pass the
# Fecha: column at the on parameter and we add the Temperatura (C) column which we add the 
# average of each one
temperature_l12w_average_per_day = temperature_l12w_df.resample('D')['Temperatura (C)'].mean().reset_index()
temperature_l12w_average_per_day = temperature_l12w_df.groupby(pd.Grouper(freq='D'))['Temperatura (C)'].mean().reset_index()
```

```python
print(temperature_l12w_average_per_day.shape)
temperature_l12w_average_per_day
```

```python
# We export the temperature_l12w_average_per_day dataframe to .csv file
temperature_l12w_average_per_day.to_csv('../../../../data/interim/Temperature/Latest_12-weeks_May-4_August-03_2018/' +'\n' 
                                  'Temperature_Average_per-day_May-4_August-03.csv', sep=',', header=True, index=False)
```
