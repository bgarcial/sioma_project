```python
from __future__ import unicode_literals

import pandas as pd
import numpy as np
```

```python
luminosity_lw12 = pd.read_csv('../../../../data/raw/FincaPorvenir/Metereologico/Latest_12-weeks_May-4_August-03_2018/' \
                                 'Luz_May-4_August-03_2018.csv', )
```

```python
print(luminosity_lw12.shape)
luminosity_lw12.head()
```

```python
print(luminosity_lw12.isnull().any())
luminosity_lw12.isnull().values.any()
```

- We going to generate in only one dataframe one record per day of luminosity values, performing an average of the values per day.

```python
# We turn to datetime the Fecha column values 
# https://pandas.pydata.org/pandas-docs/version/0.20/generated/pandas.to_datetime.html
luminosity_lw12['Fecha:'] = pd.to_datetime(luminosity_lw12['Fecha:'])

# http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.resample.html
# We use resample() function to use the data like time series per day and we pass the
# Fecha: column at the on parameter and we add the Luz (lux) column which we add the 
# average of each one

luminosity_average_per_day_l12w = luminosity_lw12.resample('D', on='Fecha:')['Luz (lux)'].mean().reset_index()
```

```python
print(luminosity_average_per_day_l12w.shape)
luminosity_average_per_day_l12w
```

```python
# We export the luminosity_average_per_day_l12w dataframe to .csv file
luminosity_average_per_day_l12w.to_csv('../../../../data/interim/Luminosity/Latest_12-weeks_May-4_August-03_2018/' +'\n' 
                                  'Luminosity_Average_per-day_May-4_August-03.csv', sep=',', header=True, index=False)
```
