- We going to generate in only one dataframe one record per day of luminosity values, performing an average of the values per day.

```python
# We turn to datetime the Fecha column values 
# https://pandas.pydata.org/pandas-docs/version/0.20/generated/pandas.to_datetime.html
luminosity['Fecha:'] = pd.to_datetime(luminosity['Fecha:'])

# http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.resample.html
# We use resample() function to use the data like time series per day and we pass the
# Fecha: column at the on parameter and we add the Luz (lux) column which we add the 
# average of each one

luminosity_average_per_day = luminosity.resample('D', on='Fecha:')['Luz (lux)'].mean().reset_index()
```

```python
print(luminosity_average_per_day.shape)
luminosity_average_per_day
```

```python
# We export the luminosity_average_per_day dataframe to .csv file
luminosity_average_per_day.to_csv('../../../../data/interim/Luminosity/28-weeks_January-12_August-03_2018/' +'\n' 
                                  'Luminosity_Average_per-day_January-12_August-03.csv', sep=',', header=True, index=False)
```
