We going to export only these columns.



```python
# http://bluegalaxy.info/codewalk/2017/06/02/pandas-how-to-export-a-dataframe-to-csv/

def partial_dataframe_to_csv(filename, DataFrame, columns):
    """Export a portion of a DataFrame to csv."""
    output = DataFrame
    # Write only the fields found in the columns list
    output = output[columns]
    output.to_csv(filename, sep=',', header=True, index=False) 

# initial_columns_to_export = ['Fecha', 'Hora', 'PORVL2N1']
```
