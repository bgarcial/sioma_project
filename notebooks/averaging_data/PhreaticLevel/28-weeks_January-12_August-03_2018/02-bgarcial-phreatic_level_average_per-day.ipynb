```python
from __future__ import unicode_literals

import pandas as pd
import numpy as np
```

# 1. Loading phreatic level datasets by lots, nodes per day values.

## 1.1  Loading L2N1

```python
phreatic_level_l2n1_28w_df = pd.read_csv('../../../../data/interim/PhreaticLevel/28-weeks_January-12_August-03_2018/' \
                                 'PhreaticLevel_By_Lots-Nodes/PORVL2N1_portion.csv', )
```

```python
print(phreatic_level_l2n1_28w_df.shape)
phreatic_level_l2n1_28w_df.head()
```

```python
# We turn to datetime the Fecha column values 
# https://pandas.pydata.org/pandas-docs/version/0.20/generated/pandas.to_datetime.html
phreatic_level_l2n1_28w_df['Fecha'] = pd.to_datetime(phreatic_level_l2n1_28w_df['Fecha'])

# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.groupby.html?highlight=groupby#pandas.DataFrame.groupby
# https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Grouper.html
# We group the dataframe phreatic_level_l2n1_28w_df the column Fecha: (se la decimos con Grouper)
# with a frecuency of D calendar days generating to each day the temperature average value
phreatic_level_l2n1_28w_average_per_day = (phreatic_level_l2n1_28w_df.groupby(pd.Grouper(key='Fecha', freq='D'))['PORVL2N1'].mean().reset_index())
```
